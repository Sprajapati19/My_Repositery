```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras import callbacks
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

### Splitting data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

### Creating model

```python
model = Sequential([
    tf.keras.Input(shape=(3,)),
    Dense(units=7, activation='relu'),
    Dense(units=7, activation = 'relu'),
    Dense(units=1, activation = 'linear')
])
```

```python
model.compile(loss= MeanSquaredError(), optimizer = 'adam')
```

```python
model.fit(X_train,y_train,epochs = 1000, callbacks = callbacks.EarlyStopping(monitor='loss',patience=15))
```

### Evaluating model

```python
#testing dataset
y_hat = model.predict(X_test)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_hat))
RMSE
```

```python
MAE = (mean_absolute_error(y_test, y_hat))
MAE
```

```python
#train dataset
y_hat_train = model.predict(X_train)
```

```python
RMSE_train = np.sqrt(mean_squared_error(y_train, y_hat_train))
RMSE_train
```

```python
MAE_train = (mean_absolute_error(y_train, y_hat_train))
MAE_train
```

### Final model

```python
final_model = Sequential([
    tf.keras.Input(shape=(3,)),
    Dense(units=7, activation='relu'),
    Dense(units=7, activation = 'relu'),
    Dense(units=1, activation = 'linear')
])
```

```python
final_model.compile(loss= MeanSquaredError(), optimizer = 'adam')
```

```python
final_model.fit(X,y,epochs = 1000, callbacks = callbacks.EarlyStopping(monitor='loss',patience=15))
```
